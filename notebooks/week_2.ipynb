```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
```

```python
print(excel_data.columns)
```

```python
# Aggregate Engagement Metrics
# Aggregating metrics for each user based on 'MSISDN/Number'
engagement_metrics = excel_data.groupby('MSISDN/Number').agg({
    'Bearer Id': 'nunique',  # Count of unique Bearer Ids as a proxy for sessions
    'Dur. (ms)': 'sum',      # Total session duration in milliseconds
    'Total DL (Bytes)': 'sum',  # Total download data in bytes
    'Total UL (Bytes)': 'sum'   # Total upload data in bytes
}).reset_index()

# Calculate total data as the sum of total download and upload data
engagement_metrics['total_data'] = engagement_metrics['Total DL (Bytes)'] + engagement_metrics['Total UL (Bytes)']
```

```python
# Identify Top Users
# Finding the top 10 users by session frequency (xDR_sessions)
top_users = engagement_metrics.nlargest(10, 'Bearer Id')
print("Top 10 Users by Session Frequency:")
print(top_users)
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans

scaler = MinMaxScaler()
# Replace 'session_duration' with 'Dur. (ms)'
normalized_data = scaler.fit_transform(engagement_metrics[['Bearer Id', 'Dur. (ms)', 'total_data']])

kmeans = KMeans(n_clusters=3, random_state=42)
engagement_metrics['cluster'] = kmeans.fit_predict(normalized_data)
```

```python
sns.scatterplot(
    x=normalized_data[:, 0], y=normalized_data[:, 1],
    hue=engagement_metrics['cluster'], palette='viridis'
)
plt.title("User Clusters")
plt.xlabel("Normalized Sessions")
plt.ylabel("Normalized Duration")
plt.show()
```

```python
# Group by clusters and compute the statistics
cluster_stats = engagement_metrics.groupby('cluster').agg({
    'Bearer Id': ['min', 'max', 'mean', 'sum'],
    'Dur. (ms)': ['min', 'max', 'mean', 'sum'],
    'total_data': ['min', 'max', 'mean', 'sum']
}).reset_index()

print("Cluster Statistics (Min, Max, Mean, Sum):")
print(cluster_stats)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Scatter plot to visualize clusters based on normalized data
sns.scatterplot(
    x=normalized_data[:, 0], y=normalized_data[:, 1],
    hue=engagement_metrics['cluster'], palette='viridis'
)
plt.title("User Engagement Clusters")
plt.xlabel("Normalized Sessions")
plt.ylabel("Normalized Duration")
plt.show()
```

```python
# Aggregate user total traffic per application
# List of application-related columns (assuming download and upload columns for each application)
applications = [
    'Social Media DL (Bytes)', 'Social Media UL (Bytes)',
    'Google DL (Bytes)', 'Google UL (Bytes)',
    'Email DL (Bytes)', 'Email UL (Bytes)',
    'Youtube DL (Bytes)', 'Youtube UL (Bytes)',
    'Netflix DL (Bytes)', 'Netflix UL (Bytes)',
    'Gaming DL (Bytes)', 'Gaming UL (Bytes)',
    'Other DL (Bytes)', 'Other UL (Bytes)'
]

# Aggregate total traffic for each application
app_usage = excel_data[applications].sum().reset_index()
app_usage.columns = ['application', 'total_data']

# Sort the applications by total data usage in descending order
app_usage = app_usage.sort_values(by='total_data', ascending=False)

# Display the top 10 applications by total data usage
print("Top 10 Applications by Total Data Usage:")
print(app_usage.head(10))

# Plot the top 3 most used applications using a bar chart
app_usage.head(3).plot(kind='bar', x='application', y='total_data', figsize=(8, 5), color='teal')
plt.title("Top 3 Applications by Data Usage")
plt.xlabel("Application")
plt.ylabel("Total Data (Bytes)")
plt.show()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Elbow Method to determine the optimal value of k
inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(normalized_data)
    inertia.append(kmeans.inertia_)

# Plot the elbow curve
plt.plot(range(1, 11), inertia, marker='o')
plt.title("Elbow Method for Optimal k")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Inertia")
plt.show()
```
